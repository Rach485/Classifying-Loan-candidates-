Now, let's make a neural network to complete the classification

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def sigmoid_derivative(sigmoid):
  return sigmoid*(1-sigmoid)

```

```python
weights0=2* np.random.random((3,10))-1
weights1=2*np.random.random((10,1))-1
```

```python
epochs=20000
learning_rate=0.01
error=[]


for epoch in range(epochs):
  input_layer=input
  sum_synapse0=input_layer.dot(weights0)
  hidden_layer=sigmoid(sum_synapse0)

  sum_synapse1=hidden_layer.dot(weights1)
  output_layer=sigmoid(sum_synapse1)

  error_output_layer=output-output_layer
  average_error=np.mean(abs(error_output_layer))

  if(epoch%1000==0):
    print("Epoch: "+str(epoch+1) + "Error: "+ str(average_error))
    error.append(average_error)
  #calculating the delta at the output layer
  delta_output_layer=error_output_layer* sigmoid_derivative(output_layer)

 # calculating delta at hidden_layer
  weights1t=weights1.T
  delta_output_layer_x_hidden_layer=delta_output_layer.dot(weights1t)
  delta_hidden_layer=delta_output_layer_x_hidden_layer*sigmoid_derivative(hidden_layer)

  # updating weights from output layer to the hidden layer
  hidden_layert=hidden_layer.T
  hidden_layer_x_delta_output_layer=hidden_layert.dot(delta_output_layer)

  weights1=weights1+(hidden_layer_x_delta_output_layer * learning_rate)

  #updating weights from the hidden layer to the output layer
  input_layert=input_layer.T
  input_layer_x_delta_hidden_layer=input_layert.dot(delta_hidden_layer)
  weighst0=weights0 + (input_layer_x_delta_hidden_layer *learning_rate)

```

Here, We developed a neuralnetwork consisting of one hidden layer and the output layer consists of one neuron.
The activation function used here is sigmoid activation function.The neural network is trained for 20000 epochs were we obtained a constant
error

```python
def calculate_output(instance):
  hidden_layer = sigmoid(np.dot(instance, weights0))
  output_layer = sigmoid(np.dot(hidden_layer, weights1))
  return output_layer[0]
```
